```python
def calculate_ATR():
    # Input values
    experience = int(input("Enter years of experience: "))
    age = int(input("Enter age: "))

    # Decision structure
    if experience > 25 and age >= 55:
        ATR = 5600000
    elif experience > 20 and age >= 45:
        ATR = 4480000
    elif experience > 10 and age >= 35:
        ATR = 1500000
    else:
        ATR = 550000

    # Output result
    print("Annual Tax Revenue (ATR) is: ₦", ATR)

# Run program
calculate_ATR()
```
