```python
import cmath

def quadratic():
    print("Quadratic Equation: Ax^2 + Bx + C = 0")
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    
    discriminant = B**2 - 4*A*C
    root1 = (-B + cmath.sqrt(discriminant)) / (2*A)
    root2 = (-B - cmath.sqrt(discriminant)) / (2*A)
    
    print("Roots are:", root1, "and", root2)

def cubic():
    print("Cubic Equation: Ax^3 + Bx^2 + Cx + D = 0")
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))
    
    # Normalize coefficients
    a = B/A
    b = C/A
    c = D/A
    
    # Depressed cubic substitution
    p = b - (a**2)/3
    q = (2*a**3)/27 - (a*b)/3 + c
    
    # Discriminant
    discriminant = (q/2)**2 + (p/3)**3
    
    if discriminant >= 0:
        u = (-q/2 + cmath.sqrt(discriminant))**(1/3)
        v = (-q/2 - cmath.sqrt(discriminant))**(1/3)
        root1 = u + v - a/3
        root2 = -(u+v)/2 - a/3 + (u-v)*cmath.sqrt(3)/2j
        root3 = -(u+v)/2 - a/3 - (u-v)*cmath.sqrt(3)/2j
        print("Roots are:", root1, root2, root3)
    else:
        r = cmath.sqrt(-(p**3)/27)
        phi = cmath.acos(-q/(2*r))
        root1 = 2*r**(1/3)*cmath.cos(phi/3) - a/3
        root2 = 2*r**(1/3)*cmath.cos((phi+2*cmath.pi)/3) - a/3
        root3 = 2*r**(1/3)*cmath.cos((phi+4*cmath.pi)/3) - a/3
        print("Roots are:", root1, root2, root3)

def quartic():
    print("Quartic Equation: Ax^4 + Bx^3 + Cx^2 + Dx + E = 0")
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))
    E = float(input("Enter E: "))
    
    # For simplicity, we reduce quartic to quadratic in x^2 if possible
    # (works when B and D are zero)
    if B == 0 and D == 0:
        # Equation looks like Ax^4 + Cx^2 + E = 0
        a = A
        b = C
        c = E
        discriminant = b**2 - 4*a*c
        y1 = (-b + cmath.sqrt(discriminant)) / (2*a)
        y2 = (-b - cmath.sqrt(discriminant)) / (2*a)
        root1 = cmath.sqrt(y1)
        root2 = -cmath.sqrt(y1)
        root3 = cmath.sqrt(y2)
        root4 = -cmath.sqrt(y2)
        print("Roots are:", root1, root2, root3, root4)
    else:
        print("General quartic solving is complex. For coursework, reduce to quadratic in x^2 if possible.")

def main():
    print("Choose the equation type:")
    print("1. Quadratic")
    print("2. Cubic")
    print("3. Quartic")
    
    choice = input("Enter choice (1/2/3): ")
    
    if choice == "1":
        quadratic()
    elif choice == "2":
        cubic()
    elif choice == "3":
        quartic()
    else:
        print("Invalid choice!")

if __name__ == "__main__":
    main()
```
