```python
import tkinter as tk
from tkinter import messagebox
import csv

def verify_employee():
    user_name = name_entry.get().strip()
    user_dept = dept_entry.get().strip()
    
    if user_name == "" or user_dept == "":
        messagebox.showwarning("Warning", "Please fill in both name and department.")
        return

    found = False
    colleagues = []

    try:
        with open('GIG-logistics.csv', mode='r') as file:
            csv_reader = csv.reader(file)
            header = next(csv_reader) 
            
            all_employees = list(csv_reader)
            
            for row in all_employees:
                if row[0].lower() == user_name.lower() and row[1].lower() == user_dept.lower():
                    found = True
                    break
            
            if found:
                for row in all_employees:
                    if row[1].lower() == user_dept.lower():
                        colleagues.append(row[0])
                
                colleagues_list = "\n".join(colleagues)
                messagebox.showinfo("Access Granted", 
                                    f"Welcome, {user_name}!\n\n"
                                    f"Members in the {user_dept} department:\n"
                                    f"{colleagues_list}")
            else:
                messagebox.showerror("Error", "Employee does not exist in our database")
                
    except FileNotFoundError:
        messagebox.showerror("File Error", "Could not find 'GIG-logistics.csv'. Make sure the file is in the same folder")

root = tk.Tk()
root.title("GIG Logistics - Identity Verification")
root.geometry("350x250")

title_label = tk.Label(root, text="Employee Verifcation System", font=("Arial", 12, "bold"))
title_label.pack(pady=10)

tk.Label(root, text="Enter Your Full Name:").pack()
name_entry = tk.Entry(root, width=35)
name_entry.pack(pady=5)

tk.Label(root, text="Enter Your Department:").pack()
dept_entry = tk.Entry(root, width=35)
dept_entry.pack(pady=5)

verify_button = tk.Button(root, text="Verify", command=verify_employee, width=15)
verify_button.pack(pady=20)

root.mainloop()
```
