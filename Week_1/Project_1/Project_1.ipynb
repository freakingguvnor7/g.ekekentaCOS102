```python
def simple_interest():
    P = float(input("Enter Principal (P): "))
    R = float(input("Enter Rate of Interest (R in %): "))
    T = float(input("Enter Time (T in years): "))
    A = P * (1 + (R/100) * T)
    print("Total Amount (Simple Interest):", A)

def compound_interest():
    P = float(input("Enter Principal (P): "))
    R = float(input("Enter Annual Rate of Interest (R in decimal, e.g. 0.05 for 5%): "))
    n = int(input("Enter number of times interest is compounded per year (n): "))
    t = float(input("Enter Time (t in years): "))
    A = P * (1 + R/n)**(n*t)
    print("Total Amount (Compound Interest):", A)

def annuity_plan():
    PMT = float(input("Enter Payment per period (PMT): "))
    R = float(input("Enter Annual Rate of Interest (R in decimal, e.g. 0.05 for 5%): "))
    n = int(input("Enter number of compounding periods per year (n): "))
    t = float(input("Enter Time (t in years): "))
    A = PMT * (((1 + R/n)**(n*t) - 1) / (R/n))
    print("Total Amount (Annuity Plan):", A)

def main():
    print("Choose the operation:")
    print("1. Simple Interest")
    print("2. Compound Interest")
    print("3. Annuity Plan")
    
    choice = input("Enter choice (1/2/3): ")
    
    if choice == "1":
        simple_interest()
    elif choice == "2":
        compound_interest()
    elif choice == "3":
        annuity_plan()
    else:
        print("Invalid choice!")

if __name__ == "__main__":
    main()
```
